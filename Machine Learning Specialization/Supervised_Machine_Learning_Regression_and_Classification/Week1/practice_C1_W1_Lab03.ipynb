```python
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])
```

```python
def f_wb(x, w, b):
    return w*x+b
```

```python
def compute_cost(x,y,w,b):
    diff_sum = 0
    m = len(x)
    for i in range(m):
        y_pred = f_wb(x[i], w, b)
        print(y_pred, y[i])
        diff_sum += (y_pred - y[i])**2
    return diff_sum/(2.0*m)
```

```python
w = 100
b = 100
print(x_train, y_train)
compute_cost(x_train, y_train, w, b)
```

```python
plt_intuition(x_train,y_train) # plt_intuition() will call compute_cost()
```
